```python
%sql
CREATE OR REPLACE TABLE high_risk_borrowers_silver AS
SELECT
  id,
  avg_cc_spending,
  total_mortgage_amount,
  has_cd_account,
  is_online_customer,
  has_securities_account,
  CASE
    WHEN avg_cc_spending > 5000 THEN 'High Spender'
    WHEN avg_cc_spending > 2000 THEN 'Medium Spender'
    ELSE 'Low Spender'
  END AS spending_category
FROM loan_details_silver
WHERE has_credit_card = true AND has_personal_loan = true
```
